```python
from deck import Deck
from hand import Hand
from card import Card
```

```python
def reshuffle_needed(deck, nb_decks, reshuffle_pct=0.9):
        return (deck.cards_left() / (52 * nb_decks)) < (reshuffle_pct)
```

```python
def is_U7(hand): 
    return 7 in [c.value for c in hand.cards]
```

```python
def is_D7(hand):
    return set([c.value for c in hand.cards]) == {7} 
```

```python
def is_T7(hand): 
    return set([c.value for c in hand.cards]) == {7}
```

```python
def is_JS(hand): 
    suits = set([c.suit for c in hand.cards])
    return (suits == {'Hearts', 'Diamonds'}) or (suits == {'Clubs', 'Spades'}) or (len(suits)==1) 
```

```python
def is_JP(hand): 
    suits = set([c.suit for c in hand.cards])
    return len(suits) == 1
```

```python
import numpy as np
from pprint import pprint

def absorption_distribution(threshold=17, min_step=1, max_step=10):
    """
    Calcule la loi de S_T via la théorie des chaînes de Markov absorbantes :
    B = (I - Q)^(-1) R
    """
    # états : 0..threshold-1 = transitoires, threshold..threshold+max_step-1 = absorbants
    transients = list(range(threshold))
    absorbings = list(range(threshold, threshold + max_step))
    
    t, a = len(transients), len(absorbings)  # nb transitoires et absorbants
    
    # Matrices Q et R
    Q = np.zeros((t, t))
    R = np.zeros((t, a))
    
    for i in transients:
        for step in range(min_step, max_step + 1):
            j = i + step
            if j < threshold:
                # transition vers un état transitoire
                Q[i, j] += 1.0 / (max_step - min_step + 1)
            else:
                # transition vers un état absorbant
                idx = absorbings.index(j)
                R[i, idx] += 1.0 / (max_step - min_step + 1)
    
    # Matrice fondamentale

    N = np.linalg.inv(np.eye(t) - Q)
    
    # Probabilités d’absorption
    B = N @ R
    
    # Loi de S_T en partant de l’état 0
    dist = {s: B[0, absorbings.index(s)] for s in absorbings}
    return dist

# Exemple d’utilisation :
dist = absorption_distribution(threshold=17, min_step=1, max_step=10)

print("Distribution de S_T (méthode matricielle) :")
pprint(dist)

def P_ge(k):
    return sum(p for s, p in dist.items() if s >= k)

print("\nProbabilités cumulées :")
print(f"P(S_T >= 18) = {P_ge(18):.6f}")
print(f"P(S_T >= 19) = {P_ge(19):.6f}")
print(f"P(S_T >= 21) = {P_ge(21):.6f}")
```
